# combined data analysis

```python
import pandas as pd

# Load all datasets
sales_pipeline = pd.read_csv('sales-pipeline.csv')
accounts = pd.read_csv('accounts.csv')
sales_teams = pd.read_csv('sales-teams.csv')
products = pd.read_csv('products.csv')
clicks = pd.read_csv('clicks.csv')

# Merge datasets for a comprehensive analysis
# Merging sales_pipeline with accounts on Account_Name and Account
merged_df = pd.merge(sales_pipeline, accounts, left_on='Account', right_on='Account', how='left')

# Merging the above with sales_teams on Sales_Agent
merged_df = pd.merge(merged_df, sales_teams, on='Sales_Agent', how='left')

# Merging the above with products on Product
merged_df = pd.merge(merged_df, products, on='Product', how='left')

# For clicks, we need a common key for a meaningful merge, which might not be directly available

# Display the head of the merged dataframe to ensure correctness
print(merged_df.head())

# Check for missing values in the merged dataframe
display(merged_df.isnull().sum())

print('Merging of datasets completed.')
```

```python
# First, let's check the columns of each dataframe to identify the correct keys for merging
print('Sales Pipeline Columns:', sales_pipeline.columns.tolist())
print('Accounts Columns:', accounts.columns.tolist())
print('Sales Teams Columns:', sales_teams.columns.tolist())
print('Products Columns:', products.columns.tolist())
print('Clicks Columns:', clicks.columns.tolist())
```

This merged dataframe now includes information from the sales pipeline, accounts, sales teams, and products datasets. It provides a comprehensive view of each sales opportunity, including the account details, sales agent information, deal stage, product details, and associated sales price.

However, there are missing values in several columns, notably in 'Account', 'Revenue', 'Employees', 'Close_Date', and 'Close_Value'. These missing values will need to be addressed during the data analysis process.

Next steps could involve analyzing sales performance, product popularity, sales team effectiveness, and account characteristics. Additionally, we can explore trends over time and identify patterns related to successful deals.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
sales_pipeline = pd.read_csv('sales-pipeline.csv')
accounts = pd.read_csv('accounts.csv')
sales_teams = pd.read_csv('sales-teams.csv')
products = pd.read_csv('products.csv')
clicks = pd.read_csv('clicks.csv')

# Visualization 1: Sales Pipeline Deal Stages
plt.figure(figsize=(10, 6))
sns.countplot(x='Deal_Stage', data=sales_pipeline, palette='Set2')
plt.title('Sales Pipeline Deal Stages')
plt.xlabel('Deal Stage')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('sales_pipeline_deal_stages.png')
plt.show()

# Visualization 2: Accounts Revenue Distribution
plt.figure(figsize=(10, 6))
sns.histplot(accounts['Revenue'], bins=30, kde=True, color='skyblue')
plt.title('Accounts Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('accounts_revenue_distribution.png')
plt.show()

# Visualization 3: Sales Teams Regional Office Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Regional_Office', data=sales_teams, palette='Set3')
plt.title('Sales Teams Regional Office Distribution')
plt.xlabel('Regional Office')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('sales_teams_regional_office_distribution.png')
plt.show()

# Visualization 4: Products Sales Price Distribution
plt.figure(figsize=(10, 6))
sns.histplot(products['Sales_Price'], bins=20, kde=True, color='purple')
plt.title('Products Sales Price Distribution')
plt.xlabel('Sales Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('products_sales_price_distribution.png')
plt.show()

# Visualization 5: Clicks Source Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Source', data=clicks, palette='Set1')
plt.title('Clicks Source Distribution')
plt.xlabel('Source')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('clicks_source_distribution.png')
plt.show()

print('Visualizations generated and saved.')
```

These visualizations provide insights into various aspects of the data, including the distribution of deal stages in the sales pipeline, revenue distribution among accounts, the distribution of sales teams across regional offices, sales price distribution for products, and the distribution of clicks by source. Each visualization helps in understanding different facets of the business operations and customer interactions.



```python
import pandas as pd

# Load the sales-pipeline data
df = pd.read_csv('sales-pipeline.csv')

# Display basic information and the first few rows of the dataframe
df_info = df.info()
df_head = df.head()

print(df_info)
print(df_head)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert 'Close_Date' and 'Created_On' to datetime
df['Close_Date'] = pd.to_datetime(df['Close_Date'])
df['Created_On'] = pd.to_datetime(df['Created_On'])

# Visualize the distribution of deals across different stages
plt.figure(figsize=(10, 6))
sns.countplot(y='Deal_Stage', data=df, order=df['Deal_Stage'].value_counts().index)
plt.title('Distribution of Deals Across Different Stages')
plt.xlabel('Count')
plt.ylabel('Deal Stage')
plt.tight_layout()
plt.show()

# Visualize the performance of sales agents
plt.figure(figsize=(12, 8))
sns.countplot(y='Sales_Agent', data=df[df['Deal_Stage'] == 'Won'], order=df[df['Deal_Stage'] == 'Won']['Sales_Agent'].value_counts().index)
plt.title('Performance of Sales Agents (Closed Deals)')
plt.xlabel('Number of Closed Deals')
plt.ylabel('Sales Agent')
plt.tight_layout()
plt.show()

# Visualize trends in close value over time
plt.figure(figsize=(12, 6))
df[df['Deal_Stage'] == 'Won'].groupby(df['Close_Date'].dt.to_period('M'))['Close_Value'].sum().plot(kind='line')
plt.title('Trends in Close Value Over Time')
plt.xlabel('Close Date')
plt.ylabel('Total Close Value')
plt.tight_layout()
plt.show()
```
